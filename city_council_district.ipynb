```python
nh = 'NEIGHBORHOOD'
def label_district(row):
    if row[nh] == 'Charlestown' or row[nh] == 'East Boston' or row[nh] == 'North End':
        return 'District 1'
    if row[nh] == 'Downtown' or row[nh] == 'South Boston' or row[nh] == 'Chinatown' or row[nh] == 'South End':
        return 'District 2'
    if row[nh] == 'Dorchester':
        return 'District 3'
    if row[nh] == 'Mattapan':
        return 'District 4'
    if row[nh] == 'Roslindale' or row[nh] == 'Hyde Park' or row[nh] == 'Readville':
        return 'District 5'
    if row[nh] == 'Jamaica Plain' or row[nh] == 'West Roxbury':
        return 'District 6'
    if row[nh] == 'Roxbury' or row[nh] == 'Roxbury Crossing':
        return 'District 7'
    if row[nh] == 'Fenway' or row[nh] == 'West End' or row[nh] == 'Beacon Hill' or row[nh] == 'Back Bay' or row[nh] == 'Kenmore':
        return 'District 8'
    if row[nh] == 'Brighton' or row[nh] == 'Allston':
        return 'District 9'
    return 'not found'

```

```python
df.apply (lambda row: label_district(row), axis=1)
```

```python
df['DISTRICT'] = df.apply (lambda row: label_district(row), axis=1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
a = df[df['Is_Affordable_Master'] == True][['DISTRICT','PID']].groupby('DISTRICT').count().reset_index()

# creating the dataset
# data = {'Affordable/Income Restricted':2579, 'Not Affordable':176019}
x = a['DISTRICT']
y = a['PID']

fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(x, y, color ='maroon',
        width = 0.4)
 
plt.xlabel("District")
plt.ylabel("Number of affordable houses")
plt.title("City council district")
plt.show()

fig.savefig('City council district.png')
```
