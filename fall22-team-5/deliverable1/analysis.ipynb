## Deliverable 1: Current Distribution of Landlords, w.r.t to affordability and Distribution characteristics of affordable housing

```python
import os

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import googlemaps
from datetime import datetime
from dotenv import load_dotenv

load_dotenv()
```

#### Step 1.1: Get General Housing Data

```python
# Read in the data
all_housing = pd.read_csv('./data/property-data.csv')
# all_housing = all_housing[['OWN_OCC', 'LU_DESC', 'UNIT_NUM', 'ZIPCODE', 'RES_UNITS']]
all_housing['ZIPCODE'] = all_housing['ZIPCODE'].apply(lambda x: str(x).zfill(5))


# Get the unique zip codes
afford_housing = pd.read_csv('./data/restricted-income-housing-2021-data.csv')
# convert Zip Code to int, if the value is NaN, leave as Nan
afford_housing['Zip Code'] = afford_housing['Zip Code'].fillna(-10).astype(np.int64)
# check if less than 5 digits, pad it with zeros
afford_housing['Zip Code'] = afford_housing['Zip Code'].apply(lambda x: str(x).zfill(5))

# explore the data
# all_housing.head()
afford_housing.head()
```

```python
# using the Google Places API to match affordable housing with an address
gmaps = googlemaps.Client(key=os.getenv('GOOGLE_API_KEY'))

def API_CALL(row):
    proj_name = row['Project Name']
    zip_code = row['Zip Code']
    places = gmaps.places(f"{proj_name} {zip_code}")
    if not places or not places['results']:
        return None
    if not "formatted_address" in places['results'][0]:
        return None
    return places['results'][0]['formatted_address']

afford_housing["formatted_address"] = afford_housing.apply(lambda x: API_CALL(x), axis=1)
afford_housing.to_csv('./data/affordable-housing-addresses.csv', index=False)
```

```python
# filter out owner occupied housing
owner_housing = all_housing[all_housing['OWN_OCC'] == 'Y']
non_owner_housing = all_housing[all_housing['OWN_OCC'] == 'N']
res_units = all_housing['RES_UNITS'].sum()
res_non_owner_units = non_owner_housing['RES_UNITS'].sum()

print("Number of owner occupied housing:", len(owner_housing))
print("Percentage of owner occupied housing:", len(owner_housing) / len(all_housing))
print("Number of non-owner occupied housing:", len(non_owner_housing))
print("Number of residential units:", res_units)
print("Number of non-owner occupied residential units:", res_non_owner_units)
print("Percentage of non-owner occupied residential units:", res_non_owner_units/res_units)
```

```python
# Plot number of RentUnits and OwnUnits by zip code and sort by max sum of RentUnits and OwnUnits
afford_housing['TotalUnits'] = afford_housing['RentUnits'] + afford_housing['OwnUnits']
afford_housing = afford_housing.sort_values(by=['TotalUnits'], ascending=False)
afford_units = afford_housing['RentUnits'].sum()
print("Number of affordable housing units:", afford_units)
print("Percentage of affordable housing units of all non-owner residential housing units:", afford_units/res_non_owner_units)
print("Percentage of affordable housing units of all residential housing units:", afford_units/res_units)
```

```python
# all housing units by zip code
owner_housing = owner_housing.groupby('ZIPCODE').sum()
owner_housing = owner_housing.reset_index()
owner_housing.head()

# affordable housing units by zip code
afford_housing = afford_housing.groupby('Zip Code').sum()
afford_housing = afford_housing.reset_index()
afford_housing.head()
```

```python
# Plot number of RentUnits and OwnUnits by zip code and sort by max sum of RentUnits and OwnUnits
afford_housing['TotalUnits'] = afford_housing['RentUnits'] + afford_housing['OwnUnits']
afford_housing = afford_housing.sort_values(by=['TotalUnits'], ascending=False)

# sum by zip code
afford_housing = afford_housing.groupby('Zip Code').sum()
afford_housing = afford_housing.reset_index()
afford_housing.head()

# sort by total units
afford_housing = afford_housing.sort_values(by=['TotalUnits'], ascending=False)

# plot RentUnits and OwnUnits by zip code on the same bar with offset to show both
plt.figure(figsize=(20,10))
plt.bar(afford_housing['Zip Code'], afford_housing['RentUnits'], bottom=afford_housing['OwnUnits'], label='RentUnits')
plt.bar(afford_housing['Zip Code'], afford_housing['OwnUnits'], label='OwnUnits')
plt.xlabel('Zip Code')
plt.ylabel('Number of Units')
plt.title('Number of RentUnits and OwnUnits by Zip Code')
plt.legend()
plt.show()

afford_housing.head()
```

### Find % of owner-occupied housing stock that is affordable housing

The following is used to generate columns so that the all_housing and afford_housing datasets can be merged.

```python
# drop all NAN formatted_address from afford_housing
afford_housing = afford_housing.dropna(subset=['formatted_address'])

# if formatted_address doesn't contain 3 commas, drop it
afford_housing = afford_housing[afford_housing['formatted_address'].str.count(',') == 3]

# if formatted_address doesn't contain a number at the start of the string, drop it (use a lambda)
afford_housing = afford_housing[afford_housing['formatted_address'].apply(lambda x: x[0].isdigit())]

# get street number for each afford_housing in dataframe using a lambda
afford_housing['ST_NUM'] = afford_housing['formatted_address'].apply(lambda x: x.split(' ')[0])

# get street name for each afford_housing in dataframe using a lambda
afford_housing['ST_NAME'] = afford_housing['formatted_address'].apply(lambda x: ' '.join(x.split(', ')[0].split(' ')[1:]))

# all caps ST_NAME
afford_housing['ST_NAME'] = afford_housing['ST_NAME'].str.upper()

# get city for each afford_housing in dataframe using a lambda
afford_housing['CITY'] = afford_housing['formatted_address'].apply(lambda x: x.split(', ')[1].split(' ')[0])

# get zipcode for each afford_housing in dataframe using a lambda
afford_housing['ZIPCODE'] = afford_housing['formatted_address'].apply(lambda x: x.split(', ')[2].split(' ')[-1])

afford_housing.head()
```

```python
import re

def st_index(row):
  st_name = row['ST_NAME']
  st_num = row['ST_NUM']

  if st_name != st_name:
    return None

  st_name = st_name.lower()
  st_name = re.sub(r'\W+', '', st_name)

  # remove suffix from street name
  st_suffix = ["ave","av", "st", "str", "street", "avenue"]
  st_name = ' '.join([word for word in st_name.split() if word not in st_suffix])

  if not st_num or st_name != st_name:
    return st_name
  
  return st_num + ' ' + st_name

afford_housing['st_index'] = afford_housing.apply(st_index, axis=1)
all_housing['st_index'] = all_housing.apply(st_index, axis=1)

afford_housing.dropna(subset=['st_index'], inplace=True)
all_housing.dropna(subset=['st_index'], inplace=True)

afford_housing['st_index'] = afford_housing['st_index'].astype(str)
all_housing['st_index'] = all_housing['st_index'].astype(str)

all_housing['IS_AFFORDABLE'] = afford_housing['st_index'].isin(all_housing['st_index'])
```

```python
# load our generated results file
overlayed_all_housing = pd.read_csv('./data/overlayed_all_housing.csv')

overlayed_all_housing.head()

print(overlayed_all_housing['IS_AFFORDABLE'].value_counts())
```

```python
# for all IS_AFFORDABLE, count the total number of units that is OWN_OCC == Y
afford_units = overlayed_all_housing[overlayed_all_housing['IS_AFFORDABLE'] == True]['OWN_OCC'].value_counts()['Y']

# find total number of OWN_OCC == Y
total_units = overlayed_all_housing['OWN_OCC'].value_counts()['Y']

print(afford_units, total_units)
```

```python
# create pie chart of own occ units that are affordable vs not affordable
labels = 'Affordable', 'Not Affordable'
sizes = [afford_units, total_units - afford_units]
explode = (0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# title
plt.title('Owner Occupied Units that are Affordable vs Not Affordable')

plt.show()
```

### find # of landlords NOT currently enrolled in affordable housing units

```python
# load our generated results file
overlayed_all_housing = pd.read_csv('./data/overlayed_all_housing.csv')

overlayed_all_housing.head()

print(overlayed_all_housing['IS_AFFORDABLE'].value_counts())
```

```python
# get all OWN_OCC == TRUE and IS_AFFORDABLE == FALSE
owner_occuiped_not_affordable = overlayed_all_housing[(overlayed_all_housing['OWN_OCC'] == 'Y') & (overlayed_all_housing['IS_AFFORDABLE'] == False)]

# get all OWN_OCC == TRUE and IS_AFFORDABLE == TRUE
owner_occuiped_affordable = overlayed_all_housing[(overlayed_all_housing['OWN_OCC'] == 'Y') & (overlayed_all_housing['IS_AFFORDABLE'] == True)]

# get all OWNERS in owner_occuiped_not_affordable but not in owner_occuiped_affordable
unenrolled_owners = owner_occuiped_not_affordable[~owner_occuiped_not_affordable['OWNER'].isin(owner_occuiped_affordable['OWNER'])]

# remove all duplicate OWNERs in unenrolled_owners
unenrolled_owners = unenrolled_owners.drop_duplicates(subset=['OWNER'])

# remove all duplicate OWNERs in owner_occuiped_affordable
enrolled_owners = owner_occuiped_affordable.drop_duplicates(subset=['OWNER'])

print('Unenrolled Owner Occupied Landlords:', len(unenrolled_owners))
print('Enrolled Owner Occupied Landlords:', len(enrolled_owners))
```

```python
# plot number of enrolled_owners vs unenrolled_owners
labels = 'Enrolled', 'Unenrolled'
sizes = [len(enrolled_owners), len(unenrolled_owners)]
explode = (0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# title
plt.title('Enrolled vs Unenrolled Owner Occupied Landlords')
plt.show()
```

```python
# get all IS_AFFORDABLE == FALSE
not_affordable = overlayed_all_housing[overlayed_all_housing['IS_AFFORDABLE'] == False]

# get all IS_AFFORDABLE == TRUE
affordable = overlayed_all_housing[overlayed_all_housing['IS_AFFORDABLE'] == True]

# get all OWNERS in not_affordable but not in affordable
unenrolled_owners = not_affordable[~not_affordable['OWNER'].isin(affordable['OWNER'])]

# remove all duplicate OWNERs in unenrolled_owners
unenrolled_owners = unenrolled_owners.drop_duplicates(subset=['OWNER'])

# remove all duplicate OWNERs in affordable
enrolled_owners = affordable.drop_duplicates(subset=['OWNER'])

print('Unenrolled Landlords:', len(unenrolled_owners))
print('Enrolled Landlords:', len(enrolled_owners))
```

```python
# plot number of enrolled_owners vs unenrolled_owners
labels = 'Enrolled', 'Unenrolled'
sizes = [len(enrolled_owners), len(unenrolled_owners)]
explode = (0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# show number of enrolled_owners vs unenrolled_owners in legend
plt.legend(['Enrolled: ' + str(len(enrolled_owners)), 'Unenrolled: ' + str(len(unenrolled_owners))])

# title
plt.title('Enrolled vs Unenrolled Landlords')
plt.show()
```
