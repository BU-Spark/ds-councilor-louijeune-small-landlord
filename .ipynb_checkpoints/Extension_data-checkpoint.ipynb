```python
import pandas as pd
```

```python
master_pd = pd.read_csv('Is_Affordable_Master_Dataset.csv')
neighborhood = pd.read_csv('boston_neighborhood.csv')
##voting_dist = pd.read_csv('boston_voting_district.csv')
area_codes = pd.read_csv('neighborhoods_and_codes.csv')
```

```python
area_codes['neighborhood']
#area_codes['area code']
area_codes['Match_Name'] = area_codes['neighborhood']
```

```python
area_codes['cleaned'] = area_codes.apply(lambda x : x['neighborhood'].split(' / ')[0],axis = 1)
area_codes['cleaned']
```

```python
def update_dict(n_dict,old_name, new_name):
    keys = [k for k, v in n_dict.items() if v == old_name]
    update_dict = dict(zip(keys,[new_name for _ in range(len(keys))]))
    
    n_dict.update(update_dict)



##update downtown
update_dict(n_dict,'Downtown Boston', 'Downtown')
```

```python
master_pd['NEIGHBORHOOD'] = master_pd.apply(lambda x : n_dict[x['ZIPCODE']]  if x['ZIPCODE'] in n_dict.keys() else 'not found', axis = 1)
```

```python
master_pd.to_csv("master_with_neighborhood_for_census.csv")
```
