```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/voting_district.csv')
df
```

```python
# Set column names to the first row values
df.columns = df.iloc[0]
df = df.drop(df.index[0])
df
```

```python
df.to_csv('../data/clean/voting_district.csv', index=False)
```
