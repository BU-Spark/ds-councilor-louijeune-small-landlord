```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/neighborhood.csv')
df
```

```python
# Set column names to the first row values
df.columns = df.iloc[0]
df = df.drop(df.index[0])
df
```

```python
df.columns
```

```python
# Rename columns
renames = {
    "field concept": "city",
    "White alone": "white",
    "Black or African American alone": "black",
    "Hispanic or Latino": "hispanic",
    "Asian, Native Hawaiian and Pacific Islander alone, all ages": "asian",
    "Other Races or Multiple Races,  all ages": "other",
    "household population": "household",
    "Institutionalized population:" : "institutionalized",
    "Correctional facilities for adults": "correctional",
    "Juvenile facilities": "juvenile",
    "Nursing facilities/Skilled-nursing facilities": "nursing",
    "Other institutional facilities": "other_institutional",
    "Noninstitutionalized population:" : "noninstitutionalized",
    "College/University student housing": "college",
    "Military quarters": "military",
    "Other noninstitutional facilities": "other_noninstitutional",
    "Occupied": "occupied",
    "Vacant": "vacant",
    "household size": "household_size",
}
df = df.rename(columns=renames)
df = df[renames.values()]
df
```

```python
df.to_csv('../data/clean/neighborhood.csv', index=False)
```
