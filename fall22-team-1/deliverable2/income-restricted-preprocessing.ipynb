```python
import pandas as pd

df = pd.read_csv('../data/income-restricted-inventory-2021.csv', dtype={'Zip Code':str})
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

rental_units = np.array(sorted(df['RentUnits'], reverse=True))
plt.title('Numbers of rental units')
plt.plot(rental_units)
```

```python
plt.title('Numbers of rental units no more than 16')
plt.plot(rental_units[rental_units<=16])
k=10
plt.axhline(k, color='r')
```

```python
df_non_comm = df[df['RentUnits']<=k]
df_non_comm = df_non_comm[df_non_comm['RentUnits'] > 0]
df_non_comm
```
